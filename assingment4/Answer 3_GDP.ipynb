```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome(r"C:\Users\DELL\Desktop\chrome\chromedriver.exe")
```

```python
driver.get("https://www.statisticstimes.com/")
driver.maximize_window()
```

```python
economy=driver.find_element(By.XPATH,"/html/body/div[2]/div[1]/div[2]/div[2]/button")
economy.click()
india=driver.find_element(By.XPATH,"/html/body/div[2]/div[1]/div[2]/div[2]/div/a[3]")
india.click()
```

```python
gdpstate=driver.find_element(By.XPATH,"/html/body/div[2]/div[2]/div[2]/ul/li[1]/a")
gdpstate.click()
```

```python
state_rank = []
state_name = []
state_gsdp1920 = []
state_gsdp1819 = []
state_share = []
state_gdp = []
```

```python
rank_tags=driver.find_elements(By.XPATH,'//div[@class="fwidth"][3]/div/div/table/tbody/tr/td[1]')
for i in rank_tags:
    rank=i.text
    state_rank.append(rank)
    
name_tags=driver.find_elements(By.XPATH,'//div[@class="fwidth"][3]/div/div/table/tbody/tr/td[2]')
for i in name_tags:
    name=i.text
    state_name.append(name)
    
gsdp1920_tags=driver.find_elements(By.XPATH,'//div[@class="fwidth"][3]/div/div/table/tbody/tr/td[3]')
for i in gsdp1920_tags:
    gsdp1920=i.text
    state_gsdp1920.append(gsdp1920)
    
gsdp1819_tags=driver.find_elements(By.XPATH,'//div[@class="fwidth"][3]/div/div/table/tbody/tr/td[4]')
for i in gsdp1819_tags:
    gsdp1819=i.text
    state_gsdp1819.append(gsdp1819)
    
share_tags=driver.find_elements(By.XPATH,'//div[@class="fwidth"][3]/div/div/table/tbody/tr/td[5]')
for i in share_tags:
    share=i.text
    state_share.append(share)
    
gdp_tags=driver.find_elements(By.XPATH,'//div[@class="fwidth"][3]/div/div/table/tbody/tr/td[6]')
for i in gdp_tags:
    gdp=i.text
    state_gdp.append(gdp)
```

```python
print(len(state_rank),len(state_name),len(state_gsdp1920),len(state_gsdp1819),len(state_share),len(state_gdp))
```

```python
import pandas as pd
df=pd.DataFrame({'State Rank':state_rank,'State Name':state_name,'State GSDP 19-20':state_gsdp1920,'State GSDP 18-19':state_gsdp1819,'State Share':state_share,'State GDP($Billion)':state_gdp})
df
```
