```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome(r"C:\Users\DELL\Desktop\chrome\chromedriver.exe")
```

```python
driver.get("https://www.imdb.com/list/ls095964455/")
driver.maximize_window()
```

```python
s_name = []
s_year = []
s_genre = []
s_time = []
s_rating = []
s_vote = []
```

```python
name_tags=driver.find_elements(By.XPATH,'//h3[@class="lister-item-header"]/a')
for i in name_tags:
        name=i.text
        s_name.append(name)
        
year_tags=driver.find_elements(By.XPATH,'//h3[@class="lister-item-header"]/span[2]')
for i in year_tags:
        year=i.text
        s_year.append(year)
        
genre_tags=driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/p/span[5]')
for i in genre_tags:
        genre=i.text
        s_genre.append(genre)
        
time_tags=driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/p/span[3]')
for i in time_tags:
        time=i.text
        s_time.append(time)
        
rating_tags=driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/div/div/span[2]')
for i in rating_tags:
        rating=i.text
        s_rating.append(rating)
        
vote_tags=driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/p[4]/span[2]')
for i in vote_tags:
        vote=i.text
        s_vote.append(vote)
```

```python
print(len(s_name),len(s_year),len(s_genre),len(s_time),len(s_rating),len(s_vote))
```

```python
import pandas as pd
df=pd.DataFrame({'Series Name':s_name,'Series Year Span':s_year,'Series Genre':s_genre,'Run Time':s_time,'Rating':s_rating,'Votes':s_vote})
df
```
