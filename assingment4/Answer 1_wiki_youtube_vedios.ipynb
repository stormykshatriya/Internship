```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome(r"C:\Users\DELL\Desktop\chrome\chromedriver.exe")
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
driver.maximize_window()
```

```python
v_rank = []
v_name = []
v_artist = []
v_views = []
v_date = []
```

```python
rank_tags=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
for i in rank_tags:
    rank=i.text
    v_rank.append(rank)
    
name_tags=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
for i in name_tags:
    name=i.text
    v_name.append(name)
    
artist_tags=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
for i in artist_tags:
    artist=i.text
    v_artist.append(artist)
    
views_tags=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
for i in views_tags:
    views=i.text
    v_views.append(views)
    
date_tags=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')
for i in date_tags:
    date=i.text
    v_date.append(date)
```

```python
print(len(v_rank),len(v_name),len(v_artist),len(v_views),len(v_date))
```

```python
import pandas as pd
df=pd.DataFrame({'Vedio Rank':v_rank,'Vedio Name':v_name,'Vedio Artist':v_artist,'Vedio Views(Billions)':v_views,'Vedio Upload Date':v_date})
df
```
