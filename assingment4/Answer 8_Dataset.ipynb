```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome(r"C:\Users\DELL\Desktop\chrome\chromedriver.exe")
```

```python
driver.get("https://archive.ics.uci.edu/ml/index.php")
driver.maximize_window()
```

```python
data=driver.find_element(By.XPATH,"/html/body/table[1]/tbody/tr/td[2]/span[2]/a/font/b")
data.click()
```

```python
ds_name = []
ds_tipe = []
ds_task = []
ds_atipe = []
ds_instances = []
ds_attributes = []
ds_year = []

name_tags=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]')
for i in name_tags:
        name=i.text
        ds_name.append(name)
        
tipe_tags=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]')
for i in tipe_tags:
        tipe=i.text
        ds_tipe.append(tipe)
        
task_tags=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]')
for i in task_tags:
        task=i.text
        ds_task.append(task)
        
atipe_tags=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]')
for i in atipe_tags:
        atipe=i.text
        ds_atipe.append(atipe)
        
instances_tags=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]')
for i in instances_tags:
        instances=i.text
        ds_instances.append(instances)
        
attributes_tags=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]')
for i in attributes_tags:
        attributes=i.text
        ds_attributes.append(attributes)
        
year_tags=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]')
for i in year_tags:
        year=i.text
        ds_year.append(year)
```

```python
print(len(ds_name),len(ds_tipe),len(ds_task),len(ds_atipe),len(ds_instances),len(ds_attributes),len(ds_year))
```

```python
import pandas as pd
df=pd.DataFrame({'Data Set Name':ds_name,'Data Type':ds_tipe,'Task':ds_task,'Attribute Type':ds_atipe,'No. Of Instances':ds_instances,'No. Of Attribute':ds_attributes,'Year':ds_year})
df
```
