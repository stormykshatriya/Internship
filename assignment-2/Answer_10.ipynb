```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome(r"C:\Users\DELL\Desktop\chrome\chromedriver.exe")
```

```python
driver.get("https://www.motor1.com/")
driver.maximize_window()
```

```python
menu=driver.find_element(By.XPATH,"/html/body/div[3]/div[2]/div/div/div[1]/div")
menu.click()
```

```python
drop=driver.find_element(By.XPATH,"/html/body/div[4]/div[1]/div[3]/ul/li[5]/button")
drop.click()
```

```python
lists=driver.find_element(By.XPATH,"/html/body/div[4]/div[1]/div[3]/ul/li[6]/ul/li[1]/a")
lists.click()
```

```python
cars=driver.find_element(By.XPATH,"/html/body/div[3]/div[8]/div[1]/div[1]/div/div/div[8]/div/div[1]/h3/a")
cars.click()
```

```python
car_name=[]
car_price=[]
```

```python
# scraping name of the car from the given page
name_tags=driver.find_elements(By.XPATH,'//h3[@class="subheader"]')
for i in name_tags[0:50]:
    name=i.text
    car_name.append(name)

# scraping prioce of the car from the given page
price_tags=driver.find_elements(By.XPATH,'//div[@class="postBody description e-content"]/p/strong')
for i in price_tags[0:50]:
    price=i.text
    car_price.append(price)
```

```python
print(len(car_name),len(car_price))
```

```python
import pandas as pd
df=pd.DataFrame({'List of Expensive Cars':car_name,'Price Of Cars':car_price})
df
```
