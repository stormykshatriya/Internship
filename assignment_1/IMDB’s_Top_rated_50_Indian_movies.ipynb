```python
# importing the required libraries 
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
name = []

for i in soup.find_all ('td',class_='titleColumn'):
    name.append(i.text.split('|')[0])
    
name
```

```python
print(len(name))
```

```python
year = []

for i in soup.find_all ('span',class_='secondaryInfo'):
    year.append(i.text)
    
year
```

```python
print(len(year))
```

```python
rating = []

for i in soup.find_all ('td',class_='ratingColumn imdbRating'):
    rating.append(i.text)
    
rating
```

```python
print(len(rating))
```

```python
# making dataframe
import pandas as pd
df = pd.DataFrame({"Movie Name":name,"Rating":rating, "Year": year} )

df[0:51]
```
