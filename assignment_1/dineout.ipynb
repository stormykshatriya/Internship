```python
# importing the required libraries 
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.dineout.co.in/lucknow-restaurants/welcome-back')
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
title = []

for i in soup.find_all ('a',class_='restnt-name ellipsis'):
    title.append(i.text.split('|')[0])
    
title
```

```python
location = []

for i in soup.find_all ('div',class_='restnt-loc ellipsis'):
    location.append(i.text.split('|')[0])
    
location
```

```python
price = []

for i in soup.find_all ('span',class_='double-line-ellipsis'):
    price.append(i.text.split('|')[0])
    
price
```

```python
image = []

for i in soup.find_all ('img',class_='no-img'):
    image.append(i.get('data-src'))
    
image    
```

```python
print(len(title),len(location),len(price),len(image))
```

```python
# making dataframe
import pandas as pd
df = pd.DataFrame({'Restaurant':title, 'Price': price, 'Location': location, 'Image URL': image})
df
```
