```python
# importing the required libraries 
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling')
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
name = []

for i in soup.find_all ('td',class_='table-body__cell rankings-table__name name'):
    name.append(i.text.split('|')[0])
    
name[0:10]
```

```python
print(len(name))
```

```python
team = []

for i in soup.find_all ('td',class_='table-body__cell nationality-logo rankings-table__team'):
    team.append(i.text.split('|')[0])
    
team[0:10]
```

```python
print(len(team))
```

```python
rank = []

for i in soup.find_all ('td',class_='table-body__cell rating'):
    rank.append(i.text.split('|')[0])
    
rank[0:10]
```

```python
print(len(rank))
```

```python
# making dataframe
import pandas as pd
df = pd.DataFrame({"Name":name,"Team":team, "Rank": rank})

df[0:10]
```
