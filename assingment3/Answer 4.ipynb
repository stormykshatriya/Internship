```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
import urllib
import csv
```

```python
driver=webdriver.Chrome(r"C:\Users\DELL\Desktop\chrome\chromedriver.exe")
```

```python
driver.get("https://www.flipkart.com/")
driver.maximize_window()
```

```python
machine_learning=driver.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input")
machine_learning.send_keys(' Oneplus Nord')
```

```python
search=driver.find_element(By.CLASS_NAME,"L0Z3Pu")
search.click()
```

```python
phn_name = []
phn_ramrom = []
phn_dsply = []
phn_camera = []
phn_batry = []
phn_wrnty = []
phn_url = []
```

```python
start=0
end=2
for page in range(start,end):
    
    name=driver.find_elements(By.XPATH,'//div[@class="_4rR01T"]')
    for i in name:
        phn_name.append(i.text)
        
        ramrom=driver.find_elements(By.XPATH,'//li[@class="rgWa7D"][1]')
    for i in ramrom:
        phn_ramrom.append(i.text)
        
        dsply=driver.find_elements(By.XPATH,'//li[@class="rgWa7D"][2]')
    for i in dsply:
        phn_dsply.append(i.text)
        
        camera=driver.find_elements(By.XPATH,'//li[@class="rgWa7D"][3]')
    for i in camera:
        phn_camera.append(i.text)
        
        batry=driver.find_elements(By.XPATH,'//li[@class="rgWa7D"][4]')
    for i in batry:
        phn_batry.append(i.text)
        
        wrnty=driver.find_elements(By.XPATH,'//li[@class="rgWa7D"][5]')
    for i in wrnty:
        phn_wrnty.append(i.text)
        
        url=driver.find_elements(By.XPATH,'//a[@class="_1fQZEK"]')
    for i in url:
        phn_url.append(i.text)
        
        
    next_button=driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]') #to scrap the data from the next page
    next_button.click()
    time.sleep(3)
```

```python
print(len(phn_name),len(phn_ramrom),len(phn_dsply),len(phn_camera),len(phn_batry),len(phn_wrnty),len(phn_url))
```

```python
import pandas as pd
df=pd.DataFrame({'Phone Name & Brand':phn_name,'RAM & ROM':phn_ramrom,'Display':phn_dsply,'Camera':phn_camera,'Battary':phn_batry,'Warranty':phn_wrnty,'URL':phn_url})
df
```
